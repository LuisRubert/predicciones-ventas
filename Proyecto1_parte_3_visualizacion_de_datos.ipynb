```python
#identificamos muchos valores faltantes en dos columnas, los cuales no tenemos manera de saber cuales son. Obtaremos por eliminar debido a que no son necesarios para la prediccion de vemtas.
columnas_a_eliminar = ['Item_Weight', 'Outlet_Size']
```

```python
df = df.drop(columns=columnas_a_eliminar)
```

# Graficos

```python
columnas_relevantes = ['Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year', 'Outlet_Location_Type', 'Outlet_Type', 'Item_Outlet_Sales']
df_relevante = df[columnas_relevantes]
```

```python
# Diagrama de dispersin para visualizar la relacin entre Item_MRP(recio Mximo Minorista (precio de lista) del producto) y Item_Outlet_Sales(Las ventas del producto en la tienda particular. Esta es la variable objetivo que se debe predecir.)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_relevante, x='Item_MRP', y='Item_Outlet_Sales')
plt.title('Relacin entre el MRP del artculo y las ventas en el punto de venta')
plt.xlabel('MRP del artculo')
plt.ylabel('Ventas en el punto de venta')
plt.show()
#Este grfico muestra la relacin entre el precio mximo de venta (MRP) de un artculo y las ventas en el punto de venta. La dispersin de puntos en el grfico nos permite observar si hay alguna tendencia o patrn entre el precio del artculo y las ventas. Si existe una relacin positiva fuerte entre el precio y las ventas, podramos inferir que los artculos ms caros tienden a venderse ms, lo que podra ayudar a predecir futuras ventas en funcin del precio.
```

```python
# Diagrama de regresin para visualizar la relacin entre Item_Visibility(El porcentaje de rea total de visualizacin de todos los productos en la tienda asignados a este producto particular) y Item_Outlet_Sales(Las ventas del producto en la tienda particular)
plt.figure(figsize=(10, 6))
sns.regplot(data=df_relevante, x='Item_Visibility', y='Item_Outlet_Sales', scatter_kws={'alpha':0.3})
plt.title('Relacin entre la visibilidad del artculo y las ventas en el punto de venta')
plt.xlabel('Visibilidad del artculo')
plt.ylabel('Ventas en el punto de venta')
plt.show()
#Este grfico muestra la relacin entre la visibilidad de un artculo en el punto de venta y las ventas en ese punto de venta. La lnea de regresin trazada en el grfico indica la tendencia general de la relacin entre la visibilidad y las ventas. Si la visibilidad tiene un impacto significativo en las ventas, podramos utilizar esta informacin para predecir las ventas futuras de un artculo basndonos en su visibilidad en el punto de venta.
```

```python
# Diagrama de barras para visualizar la relacin entre Outlet_Location_Type(El tipo de rea donde se ubica la tienda) y Item_Outlet_Sales(Las ventas del producto en la tienda particular)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_relevante, x='Outlet_Location_Type', y='Item_Outlet_Sales')
plt.title('Relacin entre el tipo de ubicacin del punto de venta y las ventas')
plt.xlabel('Tipo de ubicacin del punto de venta')
plt.ylabel('Ventas en el punto de venta')
plt.show()
#Este grfico muestra la relacin entre el tipo de ubicacin del punto de venta y las ventas en ese punto de venta. Al comparar las ventas en diferentes tipos de ubicaciones de puntos de venta, podemos identificar si hay alguna diferencia significativa en las ventas entre ellas. Esta informacin puede ayudar a comprender mejor qu factores estn contribuyendo a las diferencias en las ventas y podra utilizarse para predecir las ventas futuras en ubicaciones similares.
```

```python
# Diagrama de barras para visualizar la relacin entre Outlet_Type(Si la tienda es un almacn o algn tipo de supermercado) y Item_Outlet_Sales(Las ventas del producto en la tienda particular)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_relevante, x='Outlet_Type', y='Item_Outlet_Sales')
plt.title('Relacin entre el tipo de punto de venta y las ventas')
plt.xlabel('Tipo de punto de venta')
plt.ylabel('Ventas en el punto de venta')
plt.xticks(rotation=45)
plt.show()
#Este grfico muestra la relacin entre el tipo de punto de venta y las ventas en ese punto de venta. Al comparar las ventas en diferentes tipos de puntos de venta, como supermercados versus tiendas de conveniencia, podemos identificar si ciertos tipos de puntos de venta tienen un impacto significativo en las ventas. Esta informacin podra ser til para predecir las ventas futuras en funcin del tipo de punto de venta en el que se vende un artculo.
```
